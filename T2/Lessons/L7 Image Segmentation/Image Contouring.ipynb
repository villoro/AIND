# Image Contouring

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
image = cv2.cvtColor(cv2.imread("banana.jpg"), cv2.COLOR_BGR2RGB)
print("Type", type(image), " size:", image.shape)

plt.imshow(image)
```

### Produce the binary

```python
gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
retval, binary = cv2.threshold(gray, 245, 255, cv2.THRESH_BINARY_INV)

plt.imshow(binary, cmap="gray")
```

It is not filtering very well, we will try to transform to HSV and apply a different filter

```python
hsv = cv2.cvtColor(image, cv2.COLOR_RGB2HSV)
retval, binary = cv2.threshold(hsv[:, :, 1], 90, 255, cv2.THRESH_BINARY)

plt.imshow(binary, cmap="gray")
```

Seems to be working better

### Find and draw the contours

```python
retval, contours, hierarchy = cv2.findContours(binary, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

image_copy = np.copy(image)
all_contours = cv2.drawContours(image_copy, contours, -1, (255, 0, 0), 2)
plt.imshow(all_contours)
```
